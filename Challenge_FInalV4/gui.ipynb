```python
import requests
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. Configuraci√≥n de conexi√≥n
BASE_URL = "http://127.0.0.1:8000/api"

def call_chatbot_api(message):
    """Funci√≥n puente entre los widgets y tu API FastAPI"""
    try:
        response = requests.post(f"{BASE_URL}/chat", json={"message": message})
        response.raise_for_status()
        # Asumiendo que tu API devuelve {"response": "texto..."}
        return response.json().get("response", "Error: Sin respuesta en el JSON")
    except Exception as e:
        return f"‚ö†Ô∏è Error de conexi√≥n: {str(e)}"

# 2. Creaci√≥n de la Interfaz (Widgets)
input_box = widgets.Text(
    placeholder='Escrib√≠ tu consulta t√©cnica o cl√≠nica aqu√≠...',
    layout=widgets.Layout(width='80%')
)
send_button = widgets.Button(
    description='Consultar a Pi',
    button_style='primary', # 'success', 'info', 'warning', 'danger' or ''
    icon='paper-plane'
)
output_box = widgets.Output(
    layout={'border': '1px solid #444', 'padding': '10px', 'margin': '10px 0'}
)

# 3. L√≥gica del Chat
def on_send_button_clicked(b):
    user_message = input_box.value.strip()
    if not user_message:
        return

    with output_box:
        # Mostramos la burbuja de carga o el mensaje del usuario
        print(f"üë§ T√∫: {user_message}")
        print("ü§ñ Pi est√° pensando...")
        
        # Llamada a la API
        bot_response = call_chatbot_api(user_message)
        
        # Limpiamos solo el "pensando" y mostramos la respuesta final
        clear_output(wait=True)
        print(f"üë§ T√∫: {user_message}")
        print(f"\nü§ñ {bot_response}")
        print("-" * 50)
        
        # Limpiar el input para el pr√≥ximo mensaje
        input_box.value = ""

# 4. Vincular eventos
send_button.on_click(on_send_button_clicked)
input_box.on_submit(on_send_button_clicked) # Permite enviar con 'Enter'

# 5. Mostrar UI
print("ü¶∑ PI-EZA DENTAL - Terminal de Consulta Segura")
display(widgets.HBox([input_box, send_button]), output_box)
```
